```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

## We called matplotlib and other libraries
```

```python
%matplotlib inline

## This just works in jupyter.
## This code aim is worker than better outputs.
```

```python
## We will make an example to show any simple table.
## We will make according to year Messi's goals contribution.

## Firstly we sill make goal list and year list.

yearList = [2010, 2011, 2012, 2013, 2014, 2015]
goalList = [45, 53, 71, 60, 41, 58]

## We will make numpy aranges those lists.

npYearList = np.array(yearList)
npGoalList = np.array(goalList)

## Now we can show the table.

plt.plot(npYearList, npGoalList, "b")  ##We can make like that a table. first value is x axis, second value is y axis, third value is color of the line.
plt.xlabel("Years")  ## x axis name
plt.ylabel("Goals")  ## y axis name
plt.title("Messi's Goal Contributions According to Years")  ## Table name
plt.show()  ##For to show the scream
```

```python
## We can change some special and we can visulate again our graphs.

thisFigure = plt.figure()  ## I ensured a blank figure.
figureAxes = thisFigure.add_axes([0.2, 0.3, 0.5, 0.4])  ## 0.2 is showing us where to start this table according to x axes, 03 is showin for y axis.
                                                        ## I gave some specials. 0.5 and 0.4 are width and high of my table.
figureAxes.plot(npYearList, npGoalList, "r--")
plt.show()
```

```python
## Saving proccess

thisFigure.savefig("Messi'sGoals.png", dpi=200)
## First area is file name
## Second area is extent
```

```python
# Other graphs

## Scatter

plt.scatter(npYearList, npGoalList)
plt.show()
```

```python
## Histogram

plt.hist(npGoalList)
plt.show()

## NOTE: It can take just one dataset.
```

```python
## boxplot

plt.boxplot(npGoalList)
plt.show()
```
