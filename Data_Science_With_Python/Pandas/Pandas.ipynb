```python
# Working with Missing Datas
```

```python
## We will ensure a dict with missing datas.

weather = {"New York" : [30, 20, 25, np.nan], "London" : [15, 17, np.nan, 20], "Istanbul" : [32, 33, np.nan, np.nan], "Tokyo" : [30, 31, 33, 35]}

## np.nan, attitudes nan value.
```

```python
## Let's ensure a data frame.

weatherDataFrame = pd.DataFrame(weather)
```

```python
weatherDataFrame
```

```python
## If we don't want to work with missing values, we can eject from our list.

weatherDataFrame.dropna()
```

```python
## We can eject missing all columns.

weatherDataFrame.dropna(axis=1)
```

```python
## We can give a condition for special statuses.
## In our example we can make it, if a weather has one more than missing value, we will eject it.
weatherDataFrame.dropna(axis = 1, thresh = 3)
##If we have 3 more than not nun value, the columns won't delete.
```

```python
## Also, we can add anything for missing datas.

weatherDataFrame.fillna(27)
```

```python
# groupby
## It ensures to make groups.

## We will ensure a dataset.
## People works in a company.

companyWorkers = {"Department" : ["IT", "IT", "Law", "Law", "HR", "HR", "R-D", "R-D", "Marketing", "Marketing"],
                 "Personals" : ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"],
                 "Salary" : [1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500]}

## We will make a data frame.

companyWorkersDF = pd.DataFrame(companyWorkers)
companyWorkersDF
```

```python
## Let's make groupby

makingGroup = companyWorkersDF.groupby("Department")
makingGroup
```

```python
## We can count how many people work which departments.

makingGroup.count()
```

```python
## We can see mean

makingGroup.mean("Salary")
```

```python
## We can see all statistical values.

makingGroup.describe()
```

```python
# Merge
## We have two datasets but same datas divided two dataset.
## We can merge them.

## We have personels and in a list, their making sports.
## Second list their caloris.

sportDic = {"Names" : ["a", "b", "c", "d"], "Sport" : ["Football", "Basketball", "Volleyball", "Tennis"]}
calDic =  {"Names" : ["a", "b", "c", "d"], "Calori" : [100, 150, 200, 250]}

## We will make them data frame.
sportDF = pd.DataFrame(sportDic)
calDF = pd.DataFrame(calDic)

sportDF
calDF
```

```python
## Let's make merge

mergeTable = pd.merge(sportDF, calDF, on="Names")
mergeTable
```
